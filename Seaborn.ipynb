**Plots de distribuição**

Plots que possibilitam a visualização da distribuição dos nossos dados.

* Distplot
* Joinplot
* Pairplot
* Rugplot
* Kdeplot

```
import seaborn as sns
import numpy as np
%matplotlib inline
```

Vamos utilizar a própria fonte de dados do seaborn, o dataset tips que basicamente nos mostra dados sobre gorjetas, onde teremos informações sobre gorjetas, dia da gorjeta, se era janta ou almoço por exemplo, ect...

```
tips = sns.load_dataset('tips')
```

```
tips.info()
```

```
tips.head(5)
```

Distplot

```
sns.distplot(tips['total_bill'], kde=False, bins=50)
```

Jointplot

```
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg' )
```

Pairplot

```
sns.pairplot(tips, hue='sex')
# Matriz com todas as distribuições
```

Rugplot

```
sns.rugplot(tips['total_bill'])
# É um método de distribuição mas só faz um pequeno traço, neste exemplo podemos ver que nosso conjunto de dados está todo centrado aqui entre 10 e 20 no eixo x
```

Kdeplot

Quando nos queremos descrobrir a média ou o desvio padrão de alguns dados nem sempre esses dados vao seguir uma distruição normal mas quando calculamos esses parametros, média, desvio padrão, nós conseguimos por exemplo os melhores parametros que se encaixaria com nossos dados. O kde por exemplo é uma forma de distribuição não paramétrica, consegue encontrar uma curva de distrubuição mas nao envolve parametros, serve para dados quando voce nao tem certeza do tipo de distribuição que ele segue.

```
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

# **Plots Categóricos**

Agora vamos discutir como usar seaborn para traçar dados categóricos. Existem alguns tipos de argumentos principais para isso:

* Factorplot
* Boxplot
* Violinplot
* Stripplot
* Swarmplot
* Barplot
* Countplot


```
tips.head(5)
```

Barplot

```
sns.barplot(x='sex', y='total_bill', data=tips)
# Aqui ele está basicamente dizendo que a média da conta do total bill de homens é 21 e a linha preta especifica o desvio padrão
```

```
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
# Através do estimator=np.std, o std significa o desvio padrão, então temos o desvio padrão de 8 a 10 e pouco para os homens e desvio padrão por volta de 6,2 até 8,7 mais ou menos nas mulheres
```

Countplot

```
sns.countplot(x='sex', data=tips)
```

Boxplot

```
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
# OQUE ESTÁ ACONTECENDO AQUI????
# Boxplot basicamente plota percentis de dados ou seja basicamente o box azul a linha horizontal com vertical até a linha da primeira barra contem 25% dos dados
# da primeira linha da barra até o meio da barra 50% dos dados, do meio até final da barra 75% dos dados e da linha da ultima barra até a linha horizontal de cima é a totalidade ou seja
# 100% dos dados e os pontos acima seriam os outliers, ou seja os pontos que nao se adequam, ou estão fora do padrão.
```

```
ax = sns.boxplot(x='day', y='total_bill', data=tips, palette='Set1', hue='sex')
# Aqui por exemplo temos homem e mulher ou seja temos que as quintas feiras homens tem o total bill maior do que as mulheres e isso também vale para as sextas feiras, sabado as muheres
# tem um total bill maior e domingo os homens tem total bill maior.
```

```
sns.boxplot(data=tips, palette='rainbow', orient='h')
```

Violinplot

```
sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True)
# Aqui ele basicamente nos trouxe distribuições de probabilidade para este conjunto de dados que nos estamos vendo 
# baseado no modelo não paramétrico, no KDE, ele calcula o KDE de cada um dos dados e cria um espelho, então o violinplot
# basicamente nos mostra a distribuição dos nossos dados.
# As distribuições aqui estão divididas para homens e mulheres separadamente usando o hue.
```

Strippplot

```
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
# Então este gráfico está colocando pontos em cada contagem que ele está fazendo, poex exemplos ontem tem mais pontos em Thur.. o total bill está entre 10 e 20.
```

Swarmplot

```
# Então quando tem uma grande quantidade de pontos em um único lugar podemos usar o swarmplot para ver melhor a distribuição dos pontos
# Pode ser combinado com um Violinplot
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```
# Usando violinplot com swarmplot
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
sns.violinplot(x='day', y='total_bill', data=tips)
```

Factorplot

```
# Ele basicamente permite que criemos qualquer tipo de plot ai atribuimos um parametro kind a ele e ele assumirá a forma de cada gráfico que vimos anteriormente.
sns.factorplot(x='sex', y='total_bill', data=tips, kind='bar')
```

Exemplos:

```
sns.factorplot(x='sex', y='total_bill', data=tips, kind='box') # Oberserve o kind
```

```
sns.factorplot(x='sex', y='total_bill', data=tips, kind='violin')
```

**Plots Matriciais**

```
import seaborn as sns
%matplotlib inline
```

```
flights = sns.load_dataset('flights')
tips = sns.load_dataset('tips')
```

```
flights.head(7)
```

```
tips.head(7)
```

O principal plot matricial é o heatmap, ou mapa de calor, é importante que tanto coluna quanto indice tenham relações entre si.

```
# Temos aqui uma matriz 3 por 3 com as correlações entre os dados do conjunto de dados tips, lembrando que nas correlações só entra valores que podem ser calculados, valores numéricos, valores categóricos não entram
# Podemos ver entao neste exemplo a correlação entre total bill e total bill, total bill e tip, total bill e size, tip e total bill, tip e tip... e assim por diante.
crr = tips.corr()
crr
```

```
# Dessa forma temos a visualização desses dados, dessas correlações de tips, podemos observar que quanto mais claro maior o valor e quanto mais escuro menor o valor.
sns.heatmap(crr, annot=True) # Annot me mostra os valores
```

```
# Uma outra forma de trabalhar com heatmap é utilizando o método pivot-tables
flights.info()
```

```
# Então vamos supor que nos queremos ver a relação entre anos e meses e quantidade de passageiros.
# Podemos então utilizar pivot table para isso e usar o heatmap para ver a correlação.
pvtf = flights.pivot_table(values='passengers', index='month', columns='year')
pvtf
```

```
sns.heatmap(pvtf, linecolor='white', linewidths=1)
# Podemos observar a correlação entre mes e ano através desse mapa de calor. Lembrando que quanto mais claro mais alto a correlação.
# Oque vemos basicamente nesse plot é que quanto mais o tempo passa mais as pessoas passaram a viajar de avião.
```

```
# Clustermap basicamente vai agrupar os nossos dados em conjuntos
# Ele fez um heatmap com essas marcações indicando que esses dados dessas marcações estão agrupados de alguma forma, ele considera conjuntos relevantes, esse método serve para nos mostrar padrões nos dados.
sns.clustermap(pvtf, standard_scale=1)# Usamos esse standard scale para fazer uma "normalização dos dados" e assim conseguimos enxergar de uma forma melhor os clusters.
```

**Plots de Regressão**

```
import seaborn as sns
%matplotlib inline
```

```
tips = sns.load_dataset('tips')
```

```
# Vamos utilizar esses dados para criar modelos lineares de predição de gorjeta em função total bill
tips.head(7)
```

```
tips.info()
```

```
# Gorjeta em função do total bill, ou seja, valor total da conta.
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s':100})
# Aqui plotamos o gráfico de dispersão e ele cria um modelo linear desses dados, podemos ver na reta uma sombra que está indicando a variabilidade sendo ela naquele local de sombra uma variabilidade maior
# HUE = Segregação do conjunto de dados baseados em algum eixo categórico, no exemplo sex ele segrega o conjunto de dados em homens e mulheres
# PALETTE = Cores
# Markers = Quando temos um conjunto de dados com muitos pontos pode ser que trocar o tipo do marcador facilite a visualização
# Scatter_kws = Passo um dicionario com os valores do dicionario sendo os parametros do scatter plot que voce quer alterar, e a chave associada a esse valor. exemplo aqui, passei o parametro 's' = size e o tamanho do ponto, e assim tivemos os pontos maiores.
```

```
# Podemos também segregar esses caras, dado que por ventura tenha muitas classes diferentes no gráfico e as regressões não ficam tão claras
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col = 'sex', row='time') # Segregamos aqui com o parametro col e time.
```

```
# E para finalizar plot de regressão podemos alterar algumas dimensões dos gráficos utilizando dois parametros aspect e size
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col = 'day',aspect=0.6, size=8) 
```

**PairGrids**

```
# Com pairgrids podemos criar os mesmo gráficos que criamos anteriormente mas com um pouco mais de customização
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```
iris = sns.load_dataset('iris')
```

```
iris.head(7)
```

Iris é um dataset com espeficando tamanhos das espécies de iris.

* sepal_length = comprimento da sépala
* sepal_width = Largura da sépala
* petal_length = comprimento da sépala
* petal_width = Largura da sépala



```
iris['species'].value_counts()
# Percebemos aqui que temos 3 espécies de iris e 50 observações para cada espécie
```

```
# O pairgrid vai criar gráficos vazios e iremos manipular esses gráficos vazios
g = sns.PairGrid(iris)
# Exemplo:
g.map(plt.scatter) # Irá nos trazer esses gráficos de dispersão utilizando todas as variaveis numéricas possíveis.
```

```
# Usos do map. Não seria possível reproduzir estes gráficos com uso padrão do seaborn
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```
# Plairplot é um pairgrid feito de forma padrão
sns.pairplot(iris, hue='species')
```
