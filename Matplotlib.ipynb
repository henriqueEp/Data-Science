# **Introdução ao Matplotlib**
* Matplotlib é a biblioteca de visualização de dados mais popular do Python
* Ela te dá controle sobre todos os aspectos do plot
* Ela foi projetada para dar um sentimento semelhante aos métodos de plotagem gráfica do MatLab

```
# Importanto Bibliotecas

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
# Vamos criar um array que conta pulando do 0 a 0,5 com 11 números sendo eixo x.
x = np.linspace(0, 5, 11)
x
```

```
# Vamos criar um array que simplesmente será o eixo x ao quadrado sendo esse o eixo y.
y = x ** 2
y
```

```
# Forma Funcional
plt.plot(x, y, 'r--')
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title('Título')
```

```
# Criando plot com dois gráficos usando subplots
plt.subplots(1,2)
plt.show()
```

```
# Trabalhando com os subplots
plt.subplot(1, 2, 1) # -- Esse ultimo "1" significa que estamos trablhando com o primeiro subplot
plt.plot(x, y, 'r--') # -- 'r--' significa que queros a linha red tracejada. red = vermelho 
plt.subplot(1, 2, 2) # -- Esse ultimo "2" significa que estamos trablhando com o segundo subplot
plt.plot(x, y, 'g*-') # -- 'g*-' significa que queros a linha green tracejada. green = verde 
plt.show()
```

**Observação**

Essa é uma forma funcional e não dá total controle sobre todos os elementos da sua imagem.
O método mais seguro para se trabalhar com visualização de dados com Python é tendo conhecimento do metodo de criação de instâncias da classe matplotlib.

```
# Vamos ter um controle maior sobre essas funções
fig = plt.figure()
# Primeiro criamos uma caixa vazia com método fig.

#----------------------------------------------------#

# Vamos criar um eixo usando fig e o método add_axes
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Título')
axes2.plot(y, x)
```

```
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g-.')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Título')
```

```
fig, axs = plt.subplots(1, 2)
axs[0].plot(x, y, 'g--')
axs[0].set_title('Título')

axs[1].plot(x, y, 'r--')
axs[1].set_title('Tútulo')
```

```
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```
fig = plt.figure(figsize=(8, 4), dpi=100)
```

```
fig, axes = plt.subplots(figsize=(8,10))
axes.plot(x,y,'r--', label='x^2')
axes.plot(y,x,'g--', label='x ** 1/2')
axes.set_title('Título')
axes.legend()
```

```
fig, ax = plt.subplots(figsize=(8, 10))
ax.plot(x, x**2, 'r--', lw=10, alpha=0.5, linestyle=':')
ax.plot(x, x**3, 'b--')

ax.set_xlim([0,2])
ax.set_ylim([0,10])
```
